```python
# print Odd Numbers in the list
```

```python
list=[20,10,16,19,25,1,276,188]
print(list)
for num in list:
    if num % 2 != 0:
        print(num,"is odd")
```

```python
# print Even numbers in the list
```

```python
list=[20,10,16,19,25,1,276,188]
print(list)
for num in list:
    if num % 2 == 0:
        print(num,"is even")
```
